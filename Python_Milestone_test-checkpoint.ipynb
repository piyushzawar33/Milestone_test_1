```python
# 1.9. Write a program to find the factors of a whole number using a while loop.

def find_factors(number):
  factors = []
  divisor = 1

  while divisor <= number:
    if number % divisor == 0:
      factors.append(divisor)
    divisor += 1

  return factors

# Example usage:
number = 24
result = find_factors(number)
print(f"Factors of {number}: {result}")
```

```python
'''
1.14.Consider the gravitational interactions between the Earth, Moon, and Sun in our solar system.
Given:
mass_earth = 5.972e24 # Mass of Earth in kilograms
mass_moon = 7.34767309e22 # Mass of Moon in kilograms
mass_sun = 1.989e30 # Mass of Sun in kilograms
distance_earth_sun = 1.496e11 # Average distance between Earth and Sun in meters
distance_moon_earth = 3.844e8 # Average distance between Moon and Earth in meters
Tasks:
* Calculate the gravitational force between the Earth and the Sun.
* Calculate the gravitational force between the Moon and the Earth.
* Compare the calculated forces to determine which gravitational force is stronger.
* Explain which celestial body (Earth or Moon) is more attracted to the other based on the comparison.
'''

G = 6.67430e-11  # Gravitational constant

# Given data
mass_earth = 5.972e24
mass_moon = 7.34767309e22
mass_sun = 1.989e30
distance_earth_sun = 1.496e11
distance_moon_earth = 3.844e8

# Function to calculate gravitational force
def gravitational_force(mass1, mass2, distance):
    return G * mass1 * mass2 / (distance * distance)

# Calculate forces
force_earth_sun = gravitational_force(mass_earth, mass_sun, distance_earth_sun)
force_moon_earth = gravitational_force(mass_moon, mass_earth, distance_moon_earth)

# Compare forces
if force_earth_sun > force_moon_earth:
    stronger_force = "Earth-Sun"
    more_attracted = "Earth is more attracted to the Sun"
else:
    stronger_force = "Moon-Earth"
    more_attracted = "Moon is more attracted to the Earth"

# Print results
print("Gravitational force between Earth and Sun:", force_earth_sun, "N")
print("Gravitational force between Moon and Earth:", force_moon_earth, "N")
print("The stronger gravitational force is:", stronger_force)
print(more_attracted)
```

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance is {self.balance}.")
        else:
            print("Deposit amount must be positive.")
            
    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance is {self.balance}.")
        elif amount > self.balance:
            print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")
    
    def check_balance(self):
        print(f"Current balance: {self.balance}")

class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def calculate_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest earned: {interest}. New balance: {self.balance}")

class CheckingAccount(BankAccount):
    def __init__(self, account_number, balance, overdraft_limit):
        super().__init__(account_number, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
        else:
            print("Insufficient funds")

# Create accounts
savings_account = SavingsAccount("SA123", 1000, 0.05)
checking_account = CheckingAccount("CA456", 500, 200)

# Perform transactions
savings_account.deposit(500)
savings_account.calculate_interest()
savings_account.check_balance()

checking_account.withdraw(300)
checking_account.withdraw(800)
checking_account.check_balance()
```

```python
'''
6.Write a code for Restaurant Management System Using OOPS:
Create a MenuItem class that has attributes such as name, description, price, and category
Implement methods to add a new menu item, update menu item information, and remove a menu item from the menu
Use encapsulation to hide the menu item's unique identification number
Inherit from the MenuItem class to create a FoodItem class and a BeverageItem class, each with their own specific attributes and methods.
'''

class MenuItem:
    def __init__(self, name, description, price, category):
        self.__id = id(self)  # Encapsulated unique identifier
        self.name = name
        self.description = description
        self.price = price
        self.category = category

    def get_id(self):
        return self.__id

    def __str__(self):
        return f"ID: {self.__id}, Name: {self.name}, Price: {self.price}, Category: {self.category}"

class FoodItem(MenuItem):
    def __init__(self, name, description, price, category, is_vegetarian):
        super().__init__(name, description, price, category)
        self.is_vegetarian = is_vegetarian

class BeverageItem(MenuItem):
    def __init__(self, name, description, price, category, size):
        super().__init__(name, description, price, category)
        self.size = size

class RestaurantMenu:
    def __init__(self):
        self.menu_items = []

    def add_menu_item(self, menu_item):
        self.menu_items.append(menu_item)

    def update_menu_item(self, item_id, new_name, new_description, new_price, new_category):
        for item in self.menu_items:
            if item.get_id() == item_id:
                item.name = new_name
                item.description = new_description
                item.price = new_price
                item.category = new_category
                break

    def remove_menu_item(self, item_id):
        self.menu_items = [item for item in self.menu_items if item.get_id() != item_id]

    def print_menu(self):
        for item in self.menu_items:
            print(item)

# Example usage:
menu = RestaurantMenu()

# Add menu items
menu.add_menu_item(FoodItem("Pizza", "Delicious pizza", 15, "Main Course", False))
menu.add_menu_item(BeverageItem("Coke", "Refreshing drink", 3, "Beverages", "Regular"))

# Print menu
menu.print_menu()

# Update menu item
menu.update_menu_item(menu.menu_items[0].get_id(), "New Pizza", "Improved pizza", 18, "Main Course")

# Print menu again
menu.print_menu()

# Remove menu item
menu.remove_menu_item(menu.menu_items[1].get_id())

# Print menu after removal
menu.print_menu()
```

```python
'''
7.Write a code for Hotel Management System using OOPS:
Create a Room class that has attributes such as room number, room type, rate, and availability (private)
Implement methods to book a room, check in a guest, and check out a guest
Use encapsulation to hide the room's unique identification number.
Inherit from the Room class to create a SuiteRoom class and a StandardRoom class, each with their own specific attributes and methods.
'''

class Room:
    def __init__(self, room_number, room_type, rate):
        self.__room_number = room_number
        self.__room_type = room_type
        self.__rate = rate
        self.__available = True

    def get_room_number(self):
        return self.__room_number

    def get_room_type(self):
        return self.__room_type

    def get_rate(self):
        return self.__rate

    def is_available(self):
        return self.__available

    def book_room(self):
        if self.__available:
            self.__available = False
            print(f"Room {self.__room_number} booked successfully.")
        else:
            print(f"Room {self.__room_number} is already booked.")

    def check_in(self, guest_name):
        if self.__available:
            print(f"Guest {guest_name} checked in to room {self.__room_number}.")
            self.book_room()
        else:
            print(f"Room {self.__room_number} is not available for check-in.")

    def check_out(self):
        if not self.__available:
            self.__available = True
            print(f"Guest checked out from room {self.__room_number}.")
        else:
            print(f"Room {self.__room_number} is already vacant.")

class SuiteRoom(Room):
    def __init__(self, room_number, rate):
        super().__init__(room_number, "Suite", rate)
        self.__additional_amenities = []

    def add_amenity(self, amenity):
        self.__additional_amenities.append(amenity)

class StandardRoom(Room):
    def __init__(self, room_number, rate):
        super().__init__(room_number, "Standard", rate)

# Example usage
suite_room = SuiteRoom(101, 5000)
standard_room = StandardRoom(201, 3000)

suite_room.book_room()
suite_room.check_in("Mr. Smith")
suite_room.check_out()
standard_room.book_room()
```

```python
'''
9. Write a code for Event Management System using OOPS.

Create an Event class that has attributes such as name, date, time, location, and a list of attendees (private).
Implement methods to create a new event, add or remove attendees, and get the total number of attendees.
Use encapsulation to hide the event's unique identification number.
Inherit from the Event class to create a PrivateEvent class and a PublicEvent class, each with their own specific attributes and methods.
'''

import uuid

class Event:
    def __init__(self, name, date, time, location):
        self.__event_id = str(uuid.uuid4())
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []

    def get_event_id(self):
        return self.__event_id

    def add_attendee(self, attendee):
        self.__attendees.append(attendee)

    def remove_attendee(self, attendee):
        if attendee in self.__attendees:
            self.__attendees.remove(attendee)
            return True
        else:
            return False

    def get_attendees(self):
        return self.__attendees.copy()

    def get_total_attendees(self):
        return len(self.__attendees)

class EventManager:
    def __init__(self):
        self.events = []

    def create_event(self, name, date, time, location):
        event = Event(name, date, time, location)
        self.events.append(event)
        return event

    def get_event_by_id(self, event_id):
        for event in self.events:
            if event.get_event_id() == event_id:
                return event
        return None

    def get_all_events(self):
        return self.events

# Example usage
event_manager = EventManager()
event1 = event_manager.create_event("Concert", "2024-12-25", "20:00", "Stadium")

# Add attendee to the event directly
event1.add_attendee("John Doe")

print(event1.get_total_attendees())
```

```python
# 13. Implement a Python package structure for a project named ecommerce, containing modules for product management and order processing.

'''
Create a directory structure that reflects the package and module organization:
ecommerce/
    __init__.py
    product_management/
        __init__.py
        product.py
        inventory.py
    order_processing/
        __init__.py
        order.py
        payment.py
'''

# product.py

class Product:
    def __init__(self, product_id, name, price, stock):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.stock = stock

    def update_price(self, new_price):
        self.price = new_price

    def update_stock(self, quantity):
        self.stock += quantity

    def get_product_info(self):
        return {
            "product_id": self.product_id,
            "name": self.name,
            "price": self.price,
            "stock": self.stock
        }

# inventory.py

from .product import Product

class Inventory:
    def __init__(self):
        self.products = {}

    def add_product(self, product):
        self.products[product.product_id] = product

    def remove_product(self, product_id):
        if product_id in self.products:
            del self.products[product_id]

    def get_inventory(self):
        return {pid: p.get_product_info() for pid, p in self.products.items()}
```

```python
# 18. Create a Python script that reads a text file named "expenses.txt" and calculates the total amount spent on various expenses listed in the file.
input_file = "expenses.txt"
output_file = "total_expenses.txt"

sample_expenses = [
    15.50,
    22.30,
    10.00,
    5.75
]

# Step 1: Create the expenses.txt file with sample data
with open(input_file, "w", encoding="utf-8") as file:
    for expense in sample_expenses:
        file.write(f"{expense}\n")

print(f"File '{input_file}' has been created with sample expenses.")

# Initialize total amount
total_amount = 0.0

# Step 2: Read the expenses from the input file
try:
    with open(input_file, "r", encoding="utf-8") as file:
        for line in file:
            # Assuming each line contains a numeric value (expense amount)
            try:
                amount = float(line.strip())
                total_amount += amount
            except ValueError:
                print(f"Warning: Skipping invalid amount '{line.strip()}'")

    # Step 3: Write the total amount to the output file
    with open(output_file, "w", encoding="utf-8") as file:
        file.write(f"Total Amount Spent: ₹{total_amount:.2f}")

    print(f"Total amount spent: ₹{total_amount:.2f}")
    print(f"The total has been written to '{output_file}'.")

except FileNotFoundError:
    print(f"The file '{input_file}' does not exist.")
```

```python
# 19. Create a Python program that reads a text file named "paragraph.txt" and 
# counts the occurrences of each word in the paragraph, displaying the results in alphabetical order.

from collections import Counter
import re

# Define file names
file_name = "paragraph.txt"

# Sample content to be written to the file
sample_text = """This is a sample paragraph. It contains several words. Some words appear more than once.
The purpose is to test word counting and sorting. Enjoy counting words."""

# Step 1: Create the paragraph.txt file with sample content
with open(file_name, "w", encoding="utf-8") as file:
    file.write(sample_text)

print(f"File '{file_name}' has been created with sample content.")

# Initialize a Counter to hold word counts
word_counter = Counter()

# Step 2: Read the text file and process the words
try:
    with open(file_name, "r", encoding="utf-8") as file:
        for line in file:
            # Use regular expressions to find all words (case-insensitive)
            words = re.findall(r'\b\w+\b', line.lower())
            # Update the counter with the words from the line
            word_counter.update(words)

    # Step 3: Sort the words alphabetically and print the counts
    for word in sorted(word_counter):
        print(f"{word}: {word_counter[word]}")

except FileNotFoundError:
    print(f"The file '{file_name}' does not exist.")
```

```python
# 31. The mean of a distribution is 60 with a standard deviation of 10. Assuming that the distribution is normal,
# what percentage of items be (i) between 60 and 72, (ii) between 50 and 60, (iii) beyond 72 and (iv) between 70 and 80?

import scipy.stats as stats
mean = 60
std_dev = 10

z1 = (72 - mean) / std_dev
z2 = (50 - mean) / std_dev
z3 = (72 - mean) / std_dev
z4 = (70 - mean) / std_dev
z5 = (80 - mean) / std_dev

p1 = stats.norm.cdf(z1) - stats.norm.cdf(0)
p2 = stats.norm.cdf(0) - stats.norm.cdf(z2)
p3 = 1 - stats.norm.cdf(z3)
p4 = stats.norm.cdf(z5) - stats.norm.cdf(z4)

percent_p1 = p1 * 100
percent_p2 = p2 * 100
percent_p3 = p3 * 100
percent_p4 = p4 * 100

percent_p1, percent_p2, percent_p3, percent_p4
```

```python
# 32. 15000 students sat for an examination. The mean marks was 49 and the distribution of marks had a standard deviation of 6. 
# Assuming that the marks were normally distributed what proportion of students scored (a) more than 55 marks, (b) more than 70 marks
import numpy as np

mean_marks = 49
std_dev_marks = 6
total_students = 15000

z_55 = (55 - mean_marks) / std_dev_marks
z_70 = (70 - mean_marks) / std_dev_marks

p_55 = 1 - stats.norm.cdf(z_55)
p_70 = 1 - stats.norm.cdf(z_70)

students_more_than_55 = p_55 * total_students
students_more_than_70 = p_70 * total_students

students_more_than_55, students_more_than_70
```

```python
# 33. If the height of 500 students are normally distributed with mean 65 inch and standard deviation 5 inch.
# How many students have height : a) greater than 70 inch. b) between 60 and 70 inch.

mean_height = 65
std_dev_height = 5
total_students_height = 500

z_60 = (60 - mean_height) / std_dev_height
z_70 = (70 - mean_height) / std_dev_height

p_greater_than_70 = 1 - stats.norm.cdf(z_70)
p_between_60_and_70 = stats.norm.cdf(z_70) - stats.norm.cdf(z_60)

students_greater_than_70 = p_greater_than_70 * total_students_height
students_between_60_and_70 = p_between_60_and_70 * total_students_height

students_greater_than_70, students_between_60_and_70
```

```python
# 35.A random sample of size 25 from a population gives the sample standard derivation to be 9.0. Test the
# hypothesis that the population standard derivation is 10.5. Hint(Use chi-square distribution).

import scipy.stats as stats
n = 25
s = 9.0
sigma_0 = 10.5

chi_square_stat = (n - 1) * (s ** 2) / (sigma_0 ** 2)

df = n - 1

p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

print(f"Chi-square statistic: {chi_square_stat:.4f}")
print(f"Degrees of freedom: {df}")
print(f"P-value: {p_value:.4f}")
```

```python
# Question 37

grades = ['A', 'B', 'C', 'D', 'E']
observed_frequencies = [15, 17, 30, 22, 16]
total_students = 100
expected_frequency = total_students / len(grades)

expected_frequencies = [expected_frequency] * len(grades)

chi_square_stat, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

print(f"Chi-square statistic: {chi_square_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```
