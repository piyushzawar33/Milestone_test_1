```python
num_col = df[['AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY',
       'PEER_PRESSURE', 'CHRONIC_DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING',
       'ALCOHOL_CONSUMING', 'COUGHING', 'SHORTNESS_OF_BREATH',
       'SWALLOWING_DIFFICULTY', 'CHEST_PAIN']]

num_col
```

```python
correlation =num_col.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation,annot=True,fmt='.2f',cmap='RdBu',linewidths=0.5)
```

Doesn't seems any clear correlation.

```python
# Plotting the distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['AGE'], kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plotting the count of Lung Cancer cases
plt.figure(figsize=(8, 6))
sns.countplot(x='LUNG_CANCER', data=df, palette='Set2')
plt.title('Count of Lung Cancer Cases')
plt.xlabel('Lung Cancer')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting the relationship between Smoking, Yellow Fingers, and Lung Cancer
plt.figure(figsize=(12, 8))
sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=df, palette='coolwarm')
plt.title('Smoking vs Lung Cancer')
plt.xlabel('Smoking')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(12, 8))
sns.countplot(x='YELLOW_FINGERS', hue='LUNG_CANCER', data=df, palette='coolwarm')
plt.title('Yellow Fingers vs Lung Cancer')
plt.xlabel('Yellow Fingers')
plt.ylabel('Count')
plt.show()
```

## Insights:

### Smoking vs. Lung Cancer: 
The plot indicates a higher occurrence of lung cancer among individuals who smoke, which aligns with the known risk factors for lung cancer.

### Yellow Fingers vs. Lung Cancer: 
There is a noticeable association between the presence of yellow fingers (likely a sign of heavy smoking) and the occurrence of lung cancer, further supporting the link between smoking and lung cancer risk.

```python
# Plotting the distribution of Lung Cancer cases by Gender
plt.figure(figsize=(10, 6))
sns.countplot(x='GENDER', hue='LUNG_CANCER', data=df, palette='Set1')
plt.title('Gender Distribution in Lung Cancer Cases')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

## Insights:

The analysis reveals that lung cancer cases are almost equally distributed between males and females, with only a slight increase in incidence among males. This suggests that lung cancer risk is similar for both genders

```python
# Plotting the age distribution for lung cancer cases
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='AGE', hue='LUNG_CANCER', multiple='stack', palette='Set2', kde=True)
plt.title('Age Distribution in Lung Cancer Cases')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Insights:
Although lung cancer cases are spread across all ages, there is a slightly higher concentration of cases in individuals between the ages of 50 and 70.
